```
import pandas as pd
from nltk import ngrams
from ast import literal_eval
```

# N-grams 


```
ngram_url = 'https://raw.githubusercontent.com/manuvillalba-uclm/Natural_Lenguaje_Processing-ELM/N-grams/refined_ngram_dict.csv?token=AHJ5UMEA5POCQOL6TVGJA4273OHOG'
preprocessed_url = 'https://raw.githubusercontent.com/manuvillalba-uclm/Natural_Lenguaje_Processing-ELM/N-grams/preprocesado_v2.csv?token=AHJ5UMAKDI2JH2DZHFNY75C73PNEO'

df_ngram = pd.read_csv(ngram_url)
df_preprocessed = pd.read_csv(preprocessed_url, index_col=[0])
df_preprocessed['tweet'] = df_preprocessed['tweet'].apply(literal_eval) #reconvertir a listas la columna de tweets
#print(df_ngram.to_markdown())
#print(df_preprocessed)

#No se si puede haber un problema aqui al quitar las stopwords


# El diccionario tiene conjuntos de palabras con n = 1, 2, 3, por lo que deberemos dividirlo en tres df distintos: unigram, bigram y trigram
df_preprocessed_weights = df_preprocessed.copy()
```

```
def setWeight(frase):
  fixed_frase = " ".join(frase) #Supongo que la ultima version del preprocesado tambiÃ©n estara en vector de palabras
  found_prophate = 0
  for i in range(1,4):
    # Ejecutar la division en n-grams
    split_ngrams = ngrams(fixed_frase.split(), i)
    for gr in split_ngrams:
      posible_prophate = ' '.join(gr) #Convertimos el n-gram en frase
      ngram_search = df_ngram[df_ngram['ngram'] == posible_prophate]
      if len(ngram_search) > 0 : #Si ha encontrado alguna coincidencia, se sustituye el prophate
        aux_prophate = ngram_search['prophate'].item()
        if aux_prophate > found_prophate:
          found_prophate = aux_prophate #Nos quedamos con el mayor prophate

    # if found_prophate > 0:
    #   print('Frase: ' + fixed_frase + '\nTotal odio: ' + str(found_prophate))
  return found_prophate
```

```
df_preprocessed_weights['prophate'] = df_preprocessed_weights['tweet'].apply(setWeight)
```
